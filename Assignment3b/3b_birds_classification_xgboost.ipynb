```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('train_data.csv')
```

```python
df.head
```

```python
data=df.iloc[:, 1:]
```

```python
data
```

```python
x_train = data.iloc[:,0:9216].to_numpy()
x_train.shape
```

```python
train_labels = pd.read_csv('train_labels.csv')
```

```python
train_labels.shape
```

```python
y_train = train_labels.iloc[:,1].to_numpy()
y_train.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train = le.fit_transform(y_train)
```

```python
from xgboost import XGBClassifier
#xgbr = XGBClassifier(tree_method='gpu_hist', gpu_id=1)
xgbr = XGBClassifier(tree_method='gpu_hist',gpu_id=1,  min_child_weight = 5,learning_rate = 0.008,gamma = 0.3,colsample_bytree = 0.7)
xgbr.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

df=pd.read_csv('test_data.csv')
data=df.iloc[:, 1:]
x_test = data.iloc[:,0:9216].to_numpy()
test_labels = pd.read_csv('test_labels.csv')
y_test = test_labels.iloc[:,1].to_numpy()
y_test = le.fit_transform(y_test)
pred_labels = xgbr.predict(x_test)
accuracy = accuracy_score(y_test, pred_labels)
print('Accuracy:', accuracy)
```
