```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('bird_features.csv')
```

```python
data=df.iloc[:, 1:]
```

```python
x_train = data.iloc[:,0:9216].to_numpy()
x_train.shape
```

```python
y_train = data.iloc[:,9216].to_numpy()
y_train.shape
```

```python
#from xgboost import XGBClassifier
#xgbr = XGBClassifier(tree_method='gpu_hist', gpu_id=0)
#xgbr = XGBClassifier(tree_method='gpu_hist',gpu_id=0,  min_child_weight = 5,learning_rate = 0.008,gamma = 0.3,colsample_bytree = 0.7)
#xgbr.fit(x_train, y_train)
```

```python
from sklearn import svm
clf = svm.SVC(C=1.8, random_state=42,verbose =2)
clf.fit(x_train, y_train)
```

# Step 4: Test classification accuracy

```python
from sklearn.metrics import accuracy_score

df=pd.read_csv('bird_test.csv')
data=df.iloc[:, 1:]
x_test = data.iloc[:,0:9216].to_numpy()
y_test = data.iloc[:,9216].to_numpy()

pred_labels = clf.predict(x_test)
accuracy = accuracy_score(y_test, pred_labels)
print('Accuracy:', accuracy)
```
